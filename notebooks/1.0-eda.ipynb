```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/work_absenteeism_original.csv')
```

```python
# InformaciÃ³n bÃ¡sica
print("ðŸ”¸ Forma del dataset:", df.shape)
print("\nðŸ”¸ Columnas:", df.columns.tolist())
print("\nðŸ”¸ Tipos de datos:\n", df.dtypes)
print("\nðŸ”¸ Valores nulos por columna:")
print(df.isnull().sum())
print("\nðŸ”¸ EstadÃ­sticas descriptivas:")
display(df.describe())
```

## Visualizaciones de datos

```python
# Histograms for numerical columns
df.hist(figsize=(20, 16))
plt.show()
```

```python
# Correlation heatmap (numerical features)
fig, ax = plt.subplots(figsize=(20,16))
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', ax=ax)
plt.show()
```

```python
plt.figure(figsize=(15, len(numerical_cols)*3))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(len(numerical_cols), 1, i)
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Boxplot de {col}', fontsize=12)
    plt.xlabel('')
plt.tight_layout()
plt.show()
```
